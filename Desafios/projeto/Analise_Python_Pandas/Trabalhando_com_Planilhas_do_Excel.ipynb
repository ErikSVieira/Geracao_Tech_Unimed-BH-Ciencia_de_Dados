```python
# Importar o Pandas como pd
import pandas as pd
```

```python
# df1 = pd.read_excel('Aracaju.xlsx', sheet_name=0)  # O 'sheet_name=0' server para ler aba expecifica dentro do documeto em caso de ter varias. 
# Não esqueça de ver outras opções pois possui varias.
df1 = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/CSV_Unimed/Aracaju.xlsx')
df2 = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/CSV_Unimed/Fortaleza.xlsx')
df3 = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/CSV_Unimed/Natal.xlsx')
df4 = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/CSV_Unimed/Recife.xlsx')
df5 = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/CSV_Unimed/Salvador.xlsx')
```

```python
# Concatenar as planilhas, lembre que elas devem ser iguais
df = pd.concat([df1, df2, df3, df4, df5])
```

```python
# Exibir valores por amostragem/aleatorio
df.sample(5) # 5 tras linhas aleatorias pode ser o valor que quiser
```

```python
# Alerando o tipo de dado da coluna LojaID
df['LojaID'] = df['LojaID'].astype('object')
```

```python
df.dtypes
```

```python
df.head()
```

```python
# Consultando linhas com valores faltantes
df.isnull().sum()
```

```python
# Substituindo os valores nulos pela média
df['Vendas'].fillna(df['Vendas'].mean(), inplace=True)
# Substituindo os valores nulos por 0 (zero)
df['Vendas'].fillna(0, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.sample(15)
```

```python
# Apagar as linhas com valores nulos
df.dropna(inplace=True)
```

```python
# Apagando as linhas com base apenas em uma coluna
df.dropna(subset=['Vendas'], inplace=True)
```

```python
# Removendo linhas que estejam com valores faltantes em todas as colunas
df.dropna(how='all', inplace=True)
```

```python
df.head()
```

```python
# Criando uma coluna de Receita
df['Receita'] = df['Vendas'].mul(df['Qtde'])
# Criando um coluna de forama inversa
# df['Receita/Vendas'] = df['Receita'] / df['Vendas']
```

```python
df.head()
```

```python

df['Receita'].max()
```

```python
# nlargest
# Retornar a maior Receita com toda a linha (TOP 3)
df.nlargest(3, 'Receita')
```

```python
# Retornar a menor Receita
df['Receita'].min()
```

```python
# nsmallest
# Retornar a menor Receita com toda a linha (TOP 3)
df.nsmallest(3, 'Receita')
```

```python
# Agrupar por cidade
df.groupby('Cidade')['Receita'].sum()
```
