```python
# Mudar o tipo de dado de uma coluna
df['Idade Média'] = df['Idade Média'].astype('int')
df['Total de quartos'] = df['Total de quartos'].astype('int')
df['Total de salas'] = df['Total de salas'].astype('int')
df['População'] = df['População'].astype('int')
```

```python
# Verificar se exites valores nulos na nossa base de dados
df.isnull().sum()
```

```python
# Subistituir valores nulos na tabela
# 'inplace=True' = vai alterar as informações somente em memoria
df['nome_da_coluna_com_valores_nulos'].fillna('valor_a_ser_usado', inplace=True)
df['nome_da_coluna_com_valores_nulos'].fillna(df['Idade Média'].mean(), inplace=True)  # Pode se usar uma media da propria base de dados
df['nome_da_coluna_com_valores_nulos'].fillna(0, inplace=True)  # ou simplesmente o zero '0' 

# Apagar a linha inteira para não computar dados incompletos
df.dropna(inplace=True)

# Apagar as linhas com valores nulos com base apenas em 1 coluna
df.dropna(subset=['nome_da_coluna'], inplace=True)

# Apaga linhas que estejam com valores faltando em todas as colunas
df.dropna(how='all', inplace=True)
```

```python
# Criar novas colunas
# neste exemplo estamos pegando duas colunas um de vendas e outra de quantidade e multiplicando '.mul()' os valores
df['nome_da_nova_colua_receita'] = df['vendas_coluna_exemplo'].mul(df['quantidade_vendas_coluna_exemplo'])
# exemplo inverso
df['nome_da_nova_coluna_receita_vendas'] = df['receita_exemplo'] / df['vendas_exemplo']

# Exibir o maior valor de uma coluna
df['Receita'].max()
# Exibir o menor valor de uma coluna
df['Receita'].min()

# nlargest()
# este comando pode listar as maiores receita trazendo todas a informoções da linha encontrado
df.nlargest(3, 'Recieita')  # Este exemplo traz as 3 maiores da coluna Receita

# nsamllest
# este comando pode listar as menores receita trazendo todas a informoções da linha encontrado
df.nlargest(3, 'Recieita')  # Este exemplo traz as 3 menores da coluna Receita
```
